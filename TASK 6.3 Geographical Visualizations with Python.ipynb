```python
# Importing the gun violence file
df = pd.read_csv(os.path.join(path,'02 Data', 'Prepared Data', 'Gun_Violence_Clean_Data(2013-2018).csv'))
```

```python
# Wrangle the Data. Extract state-wise incident count
df['state'] = df['state'].astype(str)
state_incidents = df['state'].value_counts().reset_index()
state_incidents.columns = ['state', 'incident_count']
```

```python
# Clean the Data/Check for missing values
missing_values = df.isnull().sum()
```

```python
# 5. Create a Choropleth Map
# Center the map around the continental US
map_center = [37.8, -96]
choropleth_map = folium.Map(location=map_center, zoom_start=4)

folium.Choropleth(
    geo_data = country_geo,
    name = 'choropleth map',
    data = state_incidents,
    columns = ['state', 'incident_count'],
    key_on = 'feature.properties.name',
    fill_color = 'YlGnBu',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = 'Gun Violence Incidents per State (2013–2018)'
).add_to(choropleth_map)

folium.LayerControl().add_to(choropleth_map)
```

```python
# Save map as HTML in project folder
choropleth_map.save(os.path.join(path, '04 Analysis','Visualizations', "gun_violence_choropleth_map.html"))
```
